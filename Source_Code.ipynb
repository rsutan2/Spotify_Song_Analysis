```python
import numpy as np
import pandas as pd
```

```python
# Read Excel File

file = 'Spotify 2010 - 2019 Top 100 Songs.xlsx'

data = pd.ExcelFile(file)
spotify = data.parse('2010-2019')

# Change column name & print dataframe

spotify.columns=['Song_Name', 'Artist_Name', 'Song_Genre', 'Year_Released', 'Date_Added', 'BPM', 'Energy', 'Danceability', 'Decibel', 'Liveness', 'Valence', 'Duration', 'Acousticness', 'Speechness', 'Popularity_Score', 'TopYear', 'ArtistType']
spotify
```

```python
# Check column datatypes

spotify.info()
```

```python
# Move Popularity Score Column to first column

first_column = spotify.pop('Popularity_Score')
spotify.insert(0, 'Popularity_Score', first_column)

# Change Song Genre & Artist Type datatype to string

spotify['Song_Genre'] = spotify['Song_Genre'].astype('string')
spotify['ArtistType'] = spotify['ArtistType'].astype('string')
spotify
```

```python
# Change integer values for sequence purposes

def Top_Year(x):
    if x==2010: return 0
    if x==2011: return 1
    if x==2012: return 2
    if x==2013: return 3
    if x==2014: return 4
    if x==2015: return 5
    if x==2016: return 6
    if x==2017: return 7
    if x==2018: return 8
    if x==2019: return 9

spotify['Top_Year'] = spotify['TopYear'].apply(Top_Year)

# Change categorical values to numerical
    
def Artist_Type(x):
    if x=='Solo': return 1
    if x=='Duo':   return 2
    if x=='Band/Group':   return 3

spotify['Artist_Type'] = spotify['ArtistType'].apply(Artist_Type)
```

```python
# Drop irrelevant and duplicate variables

spo = spotify.drop(['Song_Name', 'Artist_Name', 'Year_Released', 'Date_Added', 'TopYear', 'ArtistType'], axis=1)
spo
```

```python
# Check column datatypes

spo.info()
```

```python
# Sort dataframe by Popularity Score in descending form

spo_final = spo.sort_values(
    by="Popularity_Score",
    ascending=False
)
spo_final.head()
```

```python
# Change to binary for popular dtype

def Popular(x):
    if x>=60: return 1
    if x<60: return 0

spo_final['Popular'] = spo_final['Popularity_Score'].apply(Popular)
```

```python
y = spo_final['Popular']
X = spo_final[['BPM', 'Energy', 'Danceability', 'Decibel', 'Liveness', 'Valence', 'Duration', 'Acousticness', 'Speechness', 'Top_Year']]
```

```python
import numpy as np
import pandas as pd

from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
models = { 
    "Linear" : LinearRegression(),
    "Logit"  : LogisticRegression(random_state=0, max_iter=1000),
    "Decision_Tree" : tree.DecisionTreeClassifier(),
    "KNN" : KNeighborsClassifier(),
    "SVM" : SVC(gamma='auto'),
}
```

```python
for name, model in models.items():
    print("Model: " + name)

    model.fit(X, y)
    pred = model.predict(X)
    print("Confusion Matrix")
    print(confusion_matrix(y, np.round(pred)))

    print("Cross val scores")
    crossvals = cross_val_score(model, X, y)
    print(cross_val_score(model, X, y))
    print("Average: " + str(crossvals.mean()))
    print()
```
